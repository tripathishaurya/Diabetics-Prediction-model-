importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

data collection and analysis

PIMA Diabetics datasets

```python
#loading the dataset to a pandas dataframe
diabetics_dataset =pd.read_csv('/content/diabetes.csv')
```

```python
#printing first five datasets
diabetics_dataset.head()
```

```python
#number of rows and columns in this dataset
diabetics_dataset.shape
```

```python
#getting the statistical measures of the data
diabetics_dataset.describe()
```

```python
diabetics_dataset['Outcome'].value_counts()
```

0--> non diabetic

1--> diabetic

```python
diabetics_dataset.groupby('Outcome').mean()
```

```python
#seperating data and labels
X= diabetics_dataset.drop(columns='Outcome',axis=1)
Y= diabetics_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

data Standirdization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=diabetics_dataset['Outcome']
```

```python
print(X)
print(Y)
```

Train Test split

```python
X_train, X_test,Y_train,Y_test =train_test_split(X,Y,test_size=.2,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the Model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

Model evaluation

Accuracy score

```python
#accuracy score on training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy score of training data :',training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('accuracy score of test data :',test_data_accuracy)
```

Making a predictive system

```python
input_data=(8,84,74,31,0,38.3,0.457,39)

#changing input_data to a numpy array
input_data_as_np_array=np.asarray(input_data)
#reashape input_data as we are prediciting it for one instance
input_data_reshaped=input_data_as_np_array.reshape(1,-1)
#standaridize the input data
std_data=scaler.transform(input_data_reshaped)
print(std_data)
prediciton=classifier.predict(std_data)
print(prediciton)
if (prediciton[0]==0):
  print('the person is diabetic')
else:
    print('the person is diabetic')
```
